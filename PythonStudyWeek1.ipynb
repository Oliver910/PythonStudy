## Day 4:迴圈(Loops)


今日任務(60分鐘)
前30分鐘:for 迴圈(和 C++ 很不同!)

````python
# C++ 風格迴圈
# for (int i = 0; i < 5; i++) {
#     cout << i;
# }

# Python 風格:使用 range()
for i in range(5):  # 0, 1, 2, 3, 4
  print(f'range(5){i}')

for i in range(2, 10):  # 2 到 9
  print(f'range(2-10){i}')

for i in range(0, 10, 2):  # 0, 2, 4, 6, 8 (步進2)
  print(f'range(0, 10, 2){i}')

# 直接迭代序列(超方便!)
fruits = ["蘋果", "香蕉", "橘子"]
for fruit in fruits:
    print(fruit)

# 需要索引時
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
````

````python
# 1. 寫一個函式判斷質數
def is_prime(n):
    for i in range(2,n):
      if n % i == 0:
        return False
    return True


# 2. 寫一個函式計算階乘
def factorial(n):
    # 你的程式碼
    result = 1
    for i in range(1,n+1):
      result *= i
    return result


# 3. 寫一個函式反轉字串
def reverse_string(text):
    # 你的程式碼
    rev_text = ""
    for i in range(len(text)-1,-1,-1):
      rev_text += text[i]
    return rev_text


# 3. 寫一個函式反轉字串
def reverse_string2(text):
  return text[::-1]

# 求 1-100間的質數
for i in range(1,101):
  if is_prime(i):
    print(f"質數:{i}")

# 求1-10 計算階乘
for i in range(1,11):
  print(f"{i}!={factorial(i)}")

print(f"ABDCDEF1234567->{reverse_string("ABDCDEF1234567")}")
print(f"1234567890->{reverse_string2("1234567890")}")
````

### 最後20分鐘:本週總複習專案

````python
# 綜合練習:學生成績管理系統
students = {}

def add_student(name, scores):
    """新增學生"""
    students[name] = scores

def get_average(name):
    """計算平均"""
    if name in students:
        return sum(students[name]) / len(students[name])
    return None

def get_top_students(n=3):
    """取得前 n 名學生"""
    averages = {name: get_average(name) for name in students}
    sorted_students = sorted(averages.items(), key=lambda x: x[1], reverse=True)
    return sorted_students[:n]

# 測試
add_student("小明", [85, 90, 78])
add_student("小華", [92, 88, 95])
add_student("小美", [76, 82, 88])

print(f"小明平均: {get_average('小明')}")
print(f"前三名: {get_top_students(3)}")
````
